**Crear un dataframe con los países longitu, latitud, regiones**

```python
countries = wb.economy.list()
df_countries = pd.DataFrame(countries)
# Generar Data Frame de los 34 paises a estudiar
df_countries = df_countries[df_countries['id'].isin(countries_list)][['id', 'value', 'region', 'longitude', 'latitude']].reset_index(drop=True)
df_countries
```

**Dataframe con códigos de los países, los indicadores en columnas y la columna time con los años**

```python
# dataframe con los indicadores como columnas y los años como filas
df_wdi = wb.data.DataFrame(indicators_list, countries_list, mrv=36, columns="series").reset_index() # most recent 5 years 
```

```python
df_wdi
```

```python
# obtener todos los códigos y los nombres de los indicadores seleccionados
cod_ind = []
name_ind = []
for indicator in indicators_list:
    ind = wb.series.get(indicator)
    cod_ind.append(ind["id"])
    name_ind.append(ind["value"])
```

```python
dict_indicators = {
    "cod_indicator": cod_ind,
    "name_indicator": name_ind
}
```

```python
df_indicators = pd.DataFrame(dict_indicators)
df_indicators
```

```python
df_indicators.to_csv("../EDA/indicadores.csv", index=False)
```
