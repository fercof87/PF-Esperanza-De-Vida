**Dataframe con las categorías de los indicadores**

En esta sección relaciono los indicadores seleccionados con el id de los tópicos(categorías)  
esto para poder relacionar los indicadores seleccionados con los tópicos(categorías), relación  
que se hace mediante el id, el código siguiente recorre todos los tópicos(categorías), y relaciona  
el id y el value(nombre) del tópico(categoría), con el código de los indicadores seleccionados para el proyecto,  
si los indicadores coinciden con los tópicos(categorías), se agregan a una lista y en otra lista el id  
del tópico(categoría)  

**Nota**  
En este proceso se pudiera almacenar en una lista el nombre del tópico(categoría), pero se decide hacerlo  
por separado, ya que el proceso es largo, son 31 indicadores, en donde cada uno será revisado para ver a  
que tópico(categoría) pertenece, y en cada tópico(categoría) hay gran cantidad de indicadores.

```python
topic_id = 1
indicators_code = []
category_id = []
i = 0
while topic_id <= 21:
    indicators_by_topic = wb.series.info(topic=topic_id)
    for item in indicators_by_topic.items:
        indicator_id = indicators_by_topic.items[i]['id']
        indicator_info = indicators_by_topic.items[i]['value']
        
        if indicator_id in indicators_list:
            indicators_code.append(indicator_id)
            category_id.append(topic_id)
        
        i += 1

    topic_id += 1
    i = 0
# print(indicators_code)
# print(category_id)
```

**Crear un diccionario con las listas para crear el dataframe**

```python
# crear diccionario
dict_category_ind = {
    "indicators_code": indicators_code,
    "category_id": category_id
}
```

**Crear el dataframe**

```python
df_indicator_category_id = pd.DataFrame(dict_category_ind)
df_indicator_category_id
```

```python
# verifico los tipos de datos, el id debe ser int
df_indicator_category_id.dtypes
```

**Crear un dataframe con todos los tópicos(categorias), con id y nombre**

En este código se extraen los id y los nombres de los tópicos(categorías)

```python
# guardar todos los tópicos en una variable
topics = {row['value']: row['id'] for row in wb.topic.list()}

# crear la lista category y llenarla con las claves
category = []
for element in topics.keys():
    category.append(element)


# crear la lista topics_id y llenarla con los ids
topic_ids = []
for value in topics.values():
    topic_ids.append(value)
```

```python
# crear un diccionario con las listas de id y category
dict_category = {
    "category_id": topic_ids,
    "category": category
}
```

**Crear el dataframe**

```python
df_category = pd.DataFrame(dict_category)
df_category
```

```python
# verificar los tipos de datos
df_category.dtypes
```

```python
# cambiar el tipo de dato de la columna id, para hacer el merge
df_category["category_id"] = pd.to_numeric(df_category["category_id"])
```

```python
# verificar el cambio
df_category.dtypes
```

```python
# unir dataframes para visualizar código indicador, el id de la categoría(tópico) y su nombre
df_indicator_category = df_indicator_category_id.merge(df_category, on="category_id")
df_indicator_category
```
