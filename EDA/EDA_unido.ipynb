```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df_pbi = pd.read_csv("Data_preparada/PBI.csv")
df_poblacion_crecimiento = pd.read_csv("Data_preparada/Data_Población_Crecimiento.csv")
df_poblacion_edad = pd.read_csv("Data_preparada/Data_Población_Edad_Genero.csv")
df_poblacion_zona = pd.read_csv("Data_preparada/Data_Población_Zona_Genero.csv")
df_esperanza = pd.read_csv("Data_preparada/Esperanza.csv")
```

```python

merged_df1_2 = pd.merge(df_pbi, df_poblacion_crecimiento, on=["país Name", "Año"], how="inner")
merged_df1_2_3 = pd.merge(merged_df1_2, df_poblacion_edad, on=["país Name", "Año"], how="inner")
merged_df1_2_3.drop(columns=["Continente_x", "Continente_y"], inplace=True)
merged_df1_2_3_4 = pd.merge(merged_df1_2_3, df_poblacion_zona, on=["país Name", "Año"], how="inner")
merged_df1_2_3_4_5 = pd.merge(merged_df1_2_3_4, df_esperanza, on=["país Name", "Año"], how="inner")
df_merge = merged_df1_2_3_4_5.drop(columns=["Continente_y", "Continente_x"])
```

```python
sns.heatmap(df_merge.corr(numeric_only=True)[["Esperanza de vida al nacer, mujeres (años)", "Esperanza de vida al nacer, total (años)", "Esperanza de vida al nacer, varones (años)"]], annot=True, cmap='coolwarm', fmt='.2f')
```
