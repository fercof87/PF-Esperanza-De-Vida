```python
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
esperanza = pd.read_excel("Data/Data_Esperanza_vida.xlsx")
esperanza
```

```python
esperanza.dropna(subset=['país Code'], inplace=True)
esperanza
```

```python
columns_to_convert = esperanza.columns[4:40]

esperanza[columns_to_convert] = esperanza[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
esperanza.info()
```

```python
esperanza.drop(columns=["2022 [YR2022]"], inplace=True)
esperanza.head(5)
```

```python
esperanza = esperanza.drop(columns=["país Code", "serie Code"])
```

```python
df_melted = esperanza.melt(id_vars=['país Name', 'serie Name'], var_name='Año', value_name='Valor')

df_pivoted = df_melted.pivot_table(index=['país Name', 'Año'], columns='serie Name', values='Valor')

df_pivoted.reset_index(inplace=True)

df_pivoted["Año"] = df_pivoted['Año'].str[:4]
df_pivoted
```

```python
df_pivoted.to_csv("Data_preparada/Esperanza.csv", index=False)
```
